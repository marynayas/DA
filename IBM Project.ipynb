### Step 3. Modeling. Create a logistic regression
model that predicts employee attrition based on the other variables available
to you. Clearly interpret your model findings. Re-run/tweak your model if you
think you can find a better fit.



```python
print(IBM_employees_copy["Attrition"].value_counts())

sns.countplot(data = IBM_employees_copy, x = "Attrition")
```

```python
# Replace "string" values to binary systaem: "Yes" = 1, "No" = 1
IBM_employees_copy["Attrition"].replace(["Yes","No"],[1, 0], inplace = True)

IBM_employees_copy["OverTime"].replace(["Yes","No"],[1, 0], inplace = True)

IBM_employees_copy["Gender"].replace(["Male","Female"],[1, 0], inplace = True)
```

```python
mod1 = sm.logit('Attrition ~ Age + Education + Gender + JobSatisfaction + MonthlyIncome + OverTime + TotalWorkingYears + YearsAtCompany + YearsInCurrentRole + YearsSinceLastPromotion', data=IBM_employees_copy).fit()

mod1.summary()
```

```python
## print just the variable coefficient values

mod1.params
```

```python
## create a dataframe with the variable names, OR's and p-values
# np.exp >> calculate the exponential of a given value

IBM_employees_copy_odds = pd.DataFrame(np.exp(mod1.params), columns = ['OR'])

## pull the pvalues directly from the model

IBM_employees_copy_odds["pvalue"] = mod1.pvalues

## create a variable to quickly determine sig

IBM_employees_copy_odds["Sig"] = np.where(IBM_employees_copy_odds["pvalue"] <= 0.05, "Sig", "Not Sig")

## print new dataset

IBM_employees_copy_odds
```
